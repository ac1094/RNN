```python
import numpy as np
import matplotlib.pyplot as plt

# Define hyperparameters
input_size = 1
hidden_size = 4
output_size = 1
learning_rate = 0.1
num_steps = 100

# Generate synthetic data (sin wave)
np.random.seed(0)
x = np.linspace(0, 10, num_steps)
y_true = np.sin(x)

# Initialize weights and biases
Wxh = np.random.randn(hidden_size, input_size) * 0.01  # input to hidden
Whh = np.random.randn(hidden_size, hidden_size) * 0.01  # hidden to hidden
Why = np.random.randn(output_size, hidden_size) * 0.01  # hidden to output
bh = np.zeros((hidden_size, 1))  # hidden bias
by = np.zeros((output_size, 1))  # output bias

# Define LSTM cell
def lstm_cell(xt, h_prev, C_prev):
    # Input gate
    it = sigmoid(np.dot(Wxh, xt) + np.dot(Whh, h_prev) + bh)
    # Forget gate
    ft = sigmoid(np.dot(Wxh, xt) + np.dot(Whh, h_prev) + bh)
    # Cell state
    Ct = ft * C_prev + it * np.tanh(np.dot(Wxh, xt) + np.dot(Whh, h_prev) + bh)
    # Output gate
    ot = sigmoid(np.dot(Wxh, xt) + np.dot(Whh, h_prev) + bh)
    # Hidden state
    ht = ot * np.tanh(Ct)
    return ht, Ct

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Training loop
losses = []
h_prev = np.zeros((hidden_size, 1))  # initial hidden state
C_prev = np.zeros((hidden_size, 1))  # initial cell state
for t in range(num_steps):
    # Forward pass
    xt = np.array([[y_true[t]]]).T  # input at time step t
    ht, Ct = lstm_cell(xt, h_prev, C_prev)
    yt_pred = np.dot(Why, ht) + by  # predicted output

    # Loss
    loss = np.mean((yt_pred - xt) ** 2)
    losses.append(loss)

    # Backward pass (not implemented for simplicity)

    # Update states
    h_prev = ht
    C_prev = Ct

    

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(x, y_true, label='True data')

# Reshape yt_pred to match the shape of x
yt_pred = np.squeeze(yt_pred)
plt.plot(x[-1]+1, yt_pred.reshape(-1,), label='Predictions')

plt.xlabel('Time Steps')
plt.ylabel('Value')
plt.title('True Data vs Predictions')
plt.legend()
plt.show()
```

```python
import numpy as np

class LSTMCell:
    def __init__(self, input_size, hidden_size):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.Wf = np.random.randn(hidden_size, input_size + hidden_size)
        self.Wi = np.random.randn(hidden_size, input_size + hidden_size)
        self.Wo = np.random.randn(hidden_size, input_size + hidden_size)
        self.Wc = np.random.randn(hidden_size, input_size + hidden_size)
        self.bf = np.random.randn(hidden_size, 1)
        self.bi = np.random.randn(hidden_size, 1)
        self.bo = np.random.randn(hidden_size, 1)
        self.bc = np.random.randn(hidden_size, 1)
        self.cell_state = np.zeros((hidden_size, 1))
        self.hidden_state = np.zeros((hidden_size, 1))
        self.dWf = np.zeros_like(self.Wf)
        self.dWi = np.zeros_like(self.Wi)
        self.dWo = np.zeros_like(self.Wo)
        self.dWc = np.zeros_like(self.Wc)
        self.dbf = np.zeros_like(self.bf)
        self.dbi = np.zeros_like(self.bi)
        self.dbo = np.zeros_like(self.bo)
        self.dbc = np.zeros_like(self.bc)

    def forward(self, x):
        concat_input = np.vstack((self.hidden_state, x))
        f = sigmoid(np.dot(self.Wf, concat_input) + self.bf)
        i = sigmoid(np.dot(self.Wi, concat_input) + self.bi)
        o = sigmoid(np.dot(self.Wo, concat_input) + self.bo)
        c_hat = np.tanh(np.dot(self.Wc, concat_input) + self.bc)

        self.cell_state = f * self.cell_state + i * c_hat
        self.hidden_state = o * np.tanh(self.cell_state)

        return self.hidden_state, self.cell_state

    def backward(self, d_hidden_state, d_cell_state):
        d_tanh_c = d_hidden_state * self.cell_state
        d_o = d_hidden_state * np.tanh(self.cell_state)
        d_cell_state += d_hidden_state * self.hidden_state * (1 - np.tanh(self.cell_state)**2)
        d_c_hat = d_cell_state * self.hidden_state
        d_f = d_cell_state * self.cell_state
        d_i = d_cell_state * d_tanh_c

        d_concat_input = np.dot(self.Wf.T, d_f) + np.dot(self.Wi.T, d_i) + np.dot(self.Wo.T, d_o) + np.dot(self.Wc.T, d_c_hat)
        self.dbf = np.sum(d_f, axis=1, keepdims=True)
        self.dbi = np.sum(d_i, axis=1, keepdims=True)
        self.dbo = np.sum(d_o, axis=1, keepdims=True)
        self.dbc = np.sum(d_c_hat, axis=1, keepdims=True)
        self.dWf = np.dot(d_f, np.vstack((self.hidden_state, x)).T)
        self.dWi = np.dot(d_i, np.vstack((self.hidden_state, x)).T)
        self.dWo = np.dot(d_o, np.vstack((self.hidden_state, x)).T)
        self.dWc = np.dot(d_c_hat, np.vstack((self.hidden_state, x)).T)

        return d_concat_input[self.hidden_size:], d_cell_state

class RNN:
    def __init__(self, input_size, hidden_size, num_layers):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm_cells = [LSTMCell(input_size, hidden_size) for _ in range(num_layers)]

    def forward(self, inputs):
        outputs = []
        for x in inputs:
            for layer in range(self.num_layers):
                hidden_state, cell_state = self.lstm_cells[layer].forward(x)
                x = hidden_state
            outputs.append(hidden_state)
        return outputs

    def backward(self, inputs, targets):
        loss = 0
        for i in range(len(inputs)):
            output = self.forward(inputs[i:i+1])[0]
            loss += np.sum((output - targets[i])**2) / 2
            d_output = output - targets[i]
            d_hidden_state = d_output
            d_cell_state = np.zeros((self.hidden_size, 1))
            for layer in reversed(range(self.num_layers)):
                d_hidden_state, d_cell_state = self.lstm_cells[layer].backward(d_hidden_state, d_cell_state)
        return loss

# Helper function for sigmoid activation
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Example usage
input_size = 2
hidden_size = 3
num_layers = 3
time_steps = 5

rnn = RNN(input_size, hidden_size, num_layers)
inputs = [np.random.randn(input_size, 1) for _ in range(time_steps)]
targets = [np.random.randn(hidden_size, 1) for _ in range(time_steps)]

loss = rnn.backward(inputs, targets)
print("Loss:", loss)
```

```python
#LSTM cell
class LSTM:
    
    def __init__(self, input_size, h_size):
        self.input_size = input_size
        self.h_size = h_size
        self.Wf = np.random.uniform(0,1,(h_size, input_size + h_size)) #Forget weights
        self.Wi = np.random.uniform(0,1,(h_size, input_size + h_size)) #Input weights
        self.Wo = np.random.uniform(0,1,(h_size, input_size + h_size)) #Output weights
        self.Wc = np.random.uniform(0,1,(h_size, input_size + h_size)) #Cell Input weights
        self.bf = np.random.randn(h_size) #Forget bias
        self.bi = np.random.randn(h_size) #Input bias
        self.bo = np.random.randn(h_size) #Output bias
        self.bc = np.random.randn(h_size) #Cell Input bias
        self.c_state = np.random.uniform(-1,1,(h_size)) #Cell state
        self.h_state = np.random.uniform(-1,1,(h_size)) #Hidden state
        
        self.Wf_g = np.zeros((h_size, input_size + h_size)) #Forget gradient
        self.Wi_g = np.zeros((h_size, input_size + h_size)) #Input Gradient
        self.Wo_g = np.zeros((h_size, input_size + h_size)) #Output Gradient
        self.Wc_g = np.zeros((h_size, input_size + h_size)) #Cell Gradient
        self.bf_g = np.zeros(h_size) #Forget bias gradient
        self.bi_g = np.zeros(h_size) #Input bias gradient
        self.bo_g = np.zeros(h_size) #Output bias gradient
        self.bc_g = np.zeros(h_size) #Cell bias Gradient
        self.c_state_g = np.zeros(h_size) #Cell State Gradient
        self.h_state_g = np.zeros(h_size) #Hidden State Gradient
        
    def display_var(self):
        print("Forward Pass:--------------------------------------------------->")
        print("Forget Weights:")
        print(self.Wf)
        print("Input Weights:")
        print(self.Wi)
        print("Output Weights:")
        print(self.Wo)
        print("Cell Input Weights:")
        print(self.Wc)
        print("Forget Bias:")
        print(self.bf)
        print("Input Bias:")
        print(self.bi)
        print("Output Bias:")
        print(self.bo)
        print("Cell Input Bias:")
        print(self.bc)
        print("Cell State:")
        print(self.c_state)
        print("Hidden State:")
        print(self.h_state)
        print("Backprop Gradients:--------------------------------------------------->")
        print("Forget Weights:")
        print(self.Wf_g)
        print("Input Weights:")
        print(self.Wi_g)
        print("Output Weights:")
        print(self.Wo_g)
        print("Cell Input Weights:")
        print(self.Wc_g)
        print("Forget Bias:")
        print(self.bf_g)
        print("Input Bias:")
        print(self.bi_g)
        print("Output Bias:")
        print(self.bo_g)
        print("Cell Input Bias:")
        print(self.bc_g)
        print("Cell State:")
        print(self.c_state_g)
        print("Hidden State:")
        print(self.h_state_g)
        
    def forward(self, x):
        concat_input = np.hstack((self.h_state, x)) #Concatenated hidden state and input
        f = sigmoid(np.dot(self.Wf, concat_input) + self.bf) #Forget cell
        i = sigmoid(np.dot(self.Wi, concat_input) + self.bi) #Input cell
        o = sigmoid(np.dot(self.Wo, concat_input) + self.bo) #Output cell
        c_hat = np.tanh(np.dot(self.Wc, concat_input) + self.bc) #Cell state

        self.c_state = f * self.c_state + i * c_hat #New Cell State
        self.h_state = o * np.tanh(self.c_state) #New Hidden State

        return self.h_state, self.c_state
    
    def backward(self, x, d_h_state, d_c_state):
        d_o = d_h_state * np.tanh(self.c_state)
        d_c_state += d_h_state * self.h_state * (1 - np.tanh(self.c_state)**2)
        concat_input = np.hstack((self.h_state, x)) #Concatenated hidden state and input
        f = sigmoid(np.dot(self.Wf, concat_input) + self.bf) #Forget cell
        d_f = d_c_state * self.c_state * f * (1-f)
        c_hat = np.tanh(np.dot(self.Wc, concat_input) + self.bc)
        i = sigmoid(np.dot(self.Wi, concat_input) + self.bi) #Input cell
        d_i = d_c_state * c_hat * i * (1-i)
        d_c_hat = d_c_state * i * (1 - c_hat**2)
        
        self.Wf_g = d_f * np.hstack((d_h_state,x)).T
        self.Wi_g = d_i * np.hstack((d_h_state,x)).T
        self.Wo_g = d_o * np.hstack((d_h_state,x)).T
        self.Wc_g = d_c_hat * np.hstack((d_h_state,x)).T
        
        self.bf_g = np.sum(d_f,axis=0,keepdims=True)
        self.bi_g = np.sum(d_f,axis=0,keepdims=True)
        self.bo_g = np.sum(d_f,axis=0,keepdims=True)
        self.bc_g = np.sum(d_f,axis=0,keepdims=True)
        
        d_concat_in = np.dot(self.Wf.T, d_f) + np.dot(self.Wi.T, d_i) + np.dot(self.Wo.T, d_o) + np.dot(self.Wc.T, d_c_hat)
        

        return d_concat_in[self.h_size:], d_c_state
```
